```python
if html.status_code == 200:
    google_soup = BeautifulSoup(html.content, "html.parser")
else:
    google_soup = html.status_code
```

```python
pymenu = google_soup.select_one('.menu').select('li')
pymenu
```

```python
for item in pymenu:
    link = item.select_one('a')
    # print(f"{link.attrs['title']} = {link.get_text()}")
    if 'title' in link.attrs:
        print(link.attrs['title'], end=" = ")
    print(link.get_text(), end=" goes to ")
    print(link.attrs['href'])
```
